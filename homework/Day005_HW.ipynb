## 作業目標

* 請利用 API: https://www.dcard.tw/_api/forums/pet/posts?popular=true 回答下列問題：

1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

```python
import requests
import json

r1 = requests.get('https://www.dcard.tw/_api/forums/pet/posts?')    #第三題使用非熱門
response1 = r1.text
data1 = json.loads(response1)
#data1

r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')   #熱門
response = r.text     #讀取文字用 .text
#response
data = json.loads(response)
data
```

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
print('這個API一次會回傳幾筆資料：',len(data),'筆')
#print(data[0])

print('-------------------------')
print('每一筆資料包含以下欄位 (方法一)：')  #https://newaurora.pixnet.net/blog/post/228249533-python-dict%E8%B3%87%E6%96%99%E8%AE%80%E5%8F%96%E3%80%81%E6%96%B0%E5%A2%9E%E3%80%81%E4%BF%AE%E6%94%B9
for i in data[0]:
    print(str(i))
    
print('-------------------------')
print('每一筆資料包含以下欄位 (方法二)：')  #https://blog.kdchang.cc/2017/08/15/learning-programming-and-coding-with-python-list-tuple-dict-set/
print(data[0].keys())     
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
for da in data:
    print('標題：',da['title'])
    print('貼文時間：',da['updatedAt'])
    print('留言人數：',da['commentCount'])
    print('按讚人數：',da['likeCount'])
    print()
    
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

print('熱門文章的「平均留言人數」與「平均按讚人數」')
countComm = 0
countLike = 0
for da in data:
    countComm += da['commentCount']
    countLike += da['likeCount']
print('平均留言人數：',countComm/len(data), ' (人)')
print('平均按讚人數：',countLike/len(data), '(人)')
print('-------------------------------------------')
print('非熱門文章的「平均留言人數」與「平均按讚人數」')
countComm1 = 0
countLike1 = 0
for da1 in data1:
    countComm1 += da1['commentCount']
    countLike1 += da1['likeCount']
print('平均留言人數：',countComm1/len(data1), ' (人)')
print('平均按讚人數：',countLike1/len(data1), '(人)')
```
